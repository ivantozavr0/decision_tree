Для каждого датасета обучим решающее дерево с параметрами по умолчанию, предварительно разбив выборку на обучающую и тестовую. Построим разделящие поверхности.

```python
def plot_surface(clf, X, y):
    plot_step = 0.01
    palette = sns.color_palette(n_colors=len(np.unique(y)))
    cmap = ListedColormap(palette)
    x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, plot_step),
                         np.arange(y_min, y_max, plot_step))
    plt.tight_layout(h_pad=0.5, w_pad=0.5, pad=2.5)

    Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)
    cs = plt.contourf(xx, yy, Z, cmap=cmap, alpha=0.3)

    plt.scatter(X[:, 0], X[:, 1], c=y, cmap=cmap, alpha=.7,
                edgecolors=np.array(palette)[y], linewidths=2)
    plt.show()
```

# Часть 2. Решающие деревья своими руками

В этой части реализуем свой класс для обучения решающего дерева в задаче бинарной классификации с возможностью обработки вещественных и категориальных признаков.

```python
import numpy as np
from collections import Counter

def H_func(branch):
    numb0 = 0
    for leaf in branch:
        numb0 += (leaf == 0)
    N = len(branch)
    p0 = numb0 / N
    p1 = 1 - p0
    return 1 - p1**2 - p0**2

def find_best_split(feature_vector, target_vector):
    """
    Под критерием Джини здесь подразумевается следующая функция:
    $$Q(R) = -\frac {|R_l|}{|R|}H(R_l) -\frac {|R_r|}{|R|}H(R_r)$$,
    $R$ — множество объектов, $R_l$ и $R_r$ — объекты, попавшие в левое и правое поддерево,
     $H(R) = 1-p_1^2-p_0^2$, $p_1$, $p_0$ — доля объектов класса 1 и 0 соответственно.

    Указания:
    * Пороги, приводящие к попаданию в одно из поддеревьев пустого множества объектов, не рассматриваются.
    * В качестве порогов, нужно брать среднее двух сосдених (при сортировке) значений признака
    * Поведение функции в случае константного признака может быть любым.
    * При одинаковых приростах Джини нужно выбирать минимальный сплит.
    * За наличие в функции циклов балл будет снижен. Векторизуйте! :)

    :param feature_vector: вещественнозначный вектор значений признака
    :param target_vector: вектор классов объектов,  len(feature_vector) == len(target_vector)

    :return thresholds: отсортированный по возрастанию вектор со всеми возможными порогами, по которым объекты можно
     разделить на две различные подвыборки, или поддерева
    :return ginis: вектор со значениями критерия Джини для каждого из порогов в thresholds len(ginis) == len(thresholds)
    :return threshold_best: оптимальный порог (число)
    :return gini_best: оптимальное значение критерия Джини (число)
    """
    # ╰( ͡° ͜ʖ ͡° )つ──☆*:・ﾟ
    if feature_vector.shape == 1 or np.all(feature_vector == feature_vector[0]):
        return (None, None, None, None)
    unity = np.vstack([feature_vector,target_vector]).T
    unity_sort = unity[unity[:, 0].argsort()]
    count = 1
    thresholds = []
    first_elem = unity_sort[0, 0]
    while count != len(feature_vector):
        threshold = (unity_sort[count-1, 0]+unity_sort[count, 0])/2
        if threshold != first_elem:
            thresholds.append(threshold)
        count += 1
    thresholds = np.array(thresholds)

    ginis = []
    r = len(feature_vector)
    for threshold in thresholds:
        barier = 0
        while (unity_sort[barier, 0]) < threshold:
            barier += 1
        r_l, r_r = barier, r - barier
        gini = -r_l / r * H_func(unity_sort[:barier, 1]) - r_r / r * H_func(unity_sort[barier:, 1])
        ginis.append(gini)
        
    gini_thresh = np.vstack([ginis,thresholds]).T
    gini_thresh = gini_thresh[gini_thresh[:, 0].argsort()]
    return (thresholds, ginis, gini_thresh[-1, 1], gini_thresh[-1, 0])
    
```

<u>Деление выборки на два поддерева следует производить на основе признака "PEG". На ее кривой существует отчетливый максимум. Кривые же для остальных признаков скорее незначительно колеблются около определенного значения коэффициента Джини. Они также имеют максимумы, которые, однако, лишь незначительно отличаются от среднего значения коэффициента Джини. Кроме того эти макисмумы гораздо меньше 0, чем максимум для "PEG". </u>

Построение дерева осуществляется согласно базовому жадному алгоритму. Выбор лучшего разбиения необходимо производить по критерию Джини. Критерий останова: все объекты в листе относятся к одному классу или ни по одному признаку нельзя разбить выборку. Ответ в листе: наиболее часто встречающийся класс в листе. Для категориальных признаков выполняется преобразование.

```python
import numpy as np
from itertools import chain


class DecisionTree(sklearn.base.BaseEstimator):
    def __init__(self, feature_types, max_depth=None, min_samples_split=None, min_samples_leaf=None):
        if np.any(list(map(lambda x: x != "real" and x != "categorical", feature_types))):
            raise ValueError("There is unknown feature type")

        self._tree = {}
        self._tree['depth'] = 0
        self.feature_types = feature_types
        self.max_depth = max_depth
        self.min_samples_split = min_samples_split
        self.min_samples_leaf = min_samples_leaf

    def _fit_node(self, sub_X, sub_y, node):
        #ош1: если в листе только один класс, то все понятно
        if np.all(sub_y == sub_y[0]):
            node["type"] = "terminal"
            node["class"] = sub_y[0]
            return

        depth = node['depth']

        if ((self.max_depth is None) or (depth < self.max_depth)) and ((self.min_samples_split is None) or len(sub_y) > self.min_samples_split):
            #будем искать по какому признаку лучше делить вершину
            feature_best, threshold_best, gini_best, split = None, None, None, None
            #ош2: нумерация с нуля
            for feature in range(sub_X.shape[1]):
                feature_type = self.feature_types[feature]
                categories_map = {}
    
                if feature_type == "real":
                    feature_vector = sub_X[:, feature]
                ##разобраться!
                elif feature_type == "categorical":
                    counts = Counter(sub_X[:, feature])
                    #ош
                    clicks = Counter(sub_X[sub_y == 1, feature])
                    ratio = {}
                    for key, current_count in counts.items():
                        if key in clicks:
                            current_click = clicks[key]
                        else:
                            current_click = 0
                        ratio[key] = current_count / current_click
                    sorted_categories = list(map(lambda x: x[1], sorted(ratio.items(), key=lambda x: x[1])))
                    categories_map = dict(zip(sorted_categories, list(range(len(sorted_categories)))))
    
                    feature_vector = np.array(map(lambda x: categories_map[x], sub_X[:, feature]))
                else:
                    raise ValueError
    
                #?? - заглушка под критерий min_samples_leaf? По умолчанию 1?
                """if len(feature_vector) == 3:
                    continue """
    
                _, _, threshold, gini = find_best_split(feature_vector, sub_y)
                if gini is None:
                    continue
                #вроде все так
                if gini_best is None or gini > gini_best:
                    feature_best = feature
                    gini_best = gini
                    split = feature_vector < threshold
    
                    if feature_type == "real":
                        threshold_best = threshold
                    #разобраться!
                    elif feature_type == "Categorical":
                        threshold_best = list(map(lambda x: x[0],
                                                  filter(lambda x: x[1] < threshold, categories_map.items())))
                    else:
                        raise ValueError
                left_leafs = sum(split)
            if (self.min_samples_leaf is not None) and (left_leafs < self.min_samples_leaf or (len(split) - left_leafs) < self.min_samples_leaf):
                feature_best = None
        else:
            feature_best = None

        if feature_best is None:
            node["type"] = "terminal"
            #ош3: нужен лишь первый элемент кортежа, являющегося первым элментом списка
            node["class"] = Counter(sub_y).most_common(1)[0][0]
            return
        
        node["type"] = "nonterminal"

        node["feature_split"] = feature_best
        if self.feature_types[feature_best] == "real":
            node["threshold"] = threshold_best
        #разобраться!
        elif self.feature_types[feature_best] == "categorical":
            node["categories_split"] = threshold_best
        else:
            raise ValueError

        #назначить левого и правого ребенка
        node["left_child"], node["right_child"] = {}, {}
        node["left_child"]['depth'], node["right_child"]['depth'] = depth + 1, depth + 1 
            
        self._fit_node(sub_X[split], sub_y[split], node["left_child"])
        #ош4, было: self._fit_node(sub_X[np.logical_not(split)], sub_y[split], node["right_child"])
        self._fit_node(sub_X[np.logical_not(split)], sub_y[np.logical_not(split)], node["right_child"])

    
    def _predict_node(self, x, node):
        # ╰( ͡° ͜ʖ ͡° )つ──☆*:・ﾟ
        if node["type"] == "terminal":
            return node["class"]
        
        feature = node["feature_split"]
        threshold = node["threshold"]
        if x[feature] < threshold:
            return self._predict_node(x, node["left_child"])
        else:
            return self._predict_node(x, node["right_child"])

    
    def fit(self, X, y):
        self._fit_node(X, y, self._tree)

    def predict(self, X):
        predicted = []
        for x in X:
            predicted.append(self._predict_node(x, self._tree))
        return np.array(predicted)
```

```python
### ╰( ͡° ͜ʖ ͡° )つ──☆*:・ﾟ
from sklearn.model_selection import cross_val_score

def cross_valid(clf, x, y):
    score = cross_val_score(clf, x, y=y, cv=10, scoring='accuracy')
    return score.mean()

data = pandas.read_csv('tic-tac-toe-endgame.csv')

target = data["V10"]
le = LabelEncoder()
target = np.array(le.fit_transform(target))

feature = np.array([[0]*(data.shape[1]-1)]*data.shape[0])
feature_types = ["real"]*feature.shape[1]

j = 0
for col in data.columns[:-1]:
    le = LabelEncoder()
    feature[:,j] = np.array(le.fit_transform(data[col]))
    j += 1

#вычисляем
clf = DecisionTree(feature_types, max_depth = 9, min_samples_split = 1, min_samples_leaf = 1)
results['LabelEncoder with my own tree'][1] = cross_valid(clf, feature, target)
```

```python
#one hot
feature = np.array([])
for col in cars.columns[:-1]:
    oh = OneHotEncoder(sparse_output=False)
    ans = np.array(oh.fit_transform(cars[col].to_numpy().reshape(-1, 1)))
    if feature.shape[0] == 0:
        feature = ans
    else:
        feature = np.hstack((feature, ans))
feature_types = ["real"]*feature.shape[1]

clf = DecisionTree(feature_types, max_depth=5, min_samples_split=5, min_samples_leaf=3)
results['OneHot with my own tree'][0] = cross_valid(clf, feature, target)
```

```python
clf = DecisionTreeClassifier(max_depth=5, min_samples_split=5, min_samples_leaf=3, random_state=13)
results['sklearn tree'][0] = cross_valid(clf, feature, target)
```
